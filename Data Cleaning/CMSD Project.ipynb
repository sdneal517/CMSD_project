```python
import pandas as pd
# !pip install xlrd

file_paths = {
    'achievement_ratings_15_16': r'C:\Users\sdnea\Documents\Data Science\CMSD Project\_Building Achievement Ratings 2015-2016.xlsx.xls',
    'achievement_ratings_16_17': r'C:\Users\sdnea\Documents\Data Science\CMSD Project\_Building Achievement Ratings 2016-2017.xlsx.xls',
    'achievement_ratings_17_18': r'C:\Users\sdnea\Documents\Data Science\CMSD Project\_Building Achievement Ratings 2017-2018.xlsx',
    'overview_17_18': r'C:\Users\sdnea\Documents\Data Science\CMSD Project\_Building Overview 2017-2018.xlsx',
    'ratings_15_16': r'C:\Users\sdnea\Documents\Data Science\CMSD Project\_Building Ratings 2015-2016.xls',
    'ratings_16_17': r'C:\Users\sdnea\Documents\Data Science\CMSD Project\_Building Ratings 2016-2017.xls',
    'value_added_15_16': r'C:\Users\sdnea\Documents\Data Science\CMSD Project\_Building Value Added 2015-2016.xls',
    'value_added_16_17': r'C:\Users\sdnea\Documents\Data Science\CMSD Project\_Building Value Added 2016-2017.xls',
    'value_added_17_18': r'C:\Users\sdnea\Documents\Data Science\CMSD Project\_Building Value Added 2017-2018.xlsx',
}

# Load each file into a DataFrame and store it in a dictionary
all_df = {name: pd.read_excel(path) for name, path in file_paths.items()}

# Iterate over the DataFrames to print column names
for name, dataframe in all_df.items():
    print(f"Columns in {name}: {dataframe.columns.tolist()}\n")

# Column names are then taken to a Google Doc (included in email) and editied to only include relevant columns
```

```python
# Define the columns to keep for each dataset
columns_to_keep = {
    'achievement_ratings_15_16': ['Building IRN', 'Building Name', 'District IRN', 'District Name', 'Address', 'City and Zip Code', 'Principal', 'Performance Index Percent 2015-16'],
    'achievement_ratings_16_17': ['Building IRN', 'Building Name', 'District IRN', 'District Name', 'Address', 'City and Zip Code', 'Principal', 'Performance Index Percent 2016-17'],
    'achievement_ratings_17_18': ['Building IRN', 'Building Name', 'District IRN', 'District Name', 'Address', 'City and Zip', 'Principal', 'Performance Index Percent 2017-18'],
    'overview_17_18': ['District IRN', 'District Name', 'Building IRN', 'Building Name', 'Address', 'City, State, Zip', 'Principal', 'Enrollment 2017-2018', 'Performance Index Percent'],
    'ratings_15_16': ['District IRN', 'District Name', 'Building IRN', 'Building Name', 'Address', 'City, State, Zip', 'Principal', 'Enrollment 2015-2016', 'Performance Index Percent', 'Letter Grade of Overall Value Added'],
    'ratings_16_17': ['District IRN', 'District Name', 'Building IRN', 'Building Name', 'Address', 'City, State, Zip', 'Principal', 'Enrollment 2016-2017', 'Performance Index Percent'],
    'value_added_15_16': ['District IRN', 'District Name', 'Building IRN', 'Building Name', 'Overall Value Added Grade'],
    'value_added_16_17': ['District IRN', 'District Name', 'Building IRN', 'Building Name', 'Overall Value Added Grade'],
    'value_added_17_18': ['District IRN', 'District Name', 'Building IRN', 'Building Name', 'Overall Value Added Grade']
}

# Update each DataFrame in the all_df dictionary to only keep specified columns
for name, dataframe in all_df.items():
    all_df[name] = dataframe.loc[:, columns_to_keep[name]]

# Verify only the desired columns are left
for name, dataframe in all_df.items():
    print(f"Columns in {name} after update: {dataframe.columns.tolist()}\n")
```

```python
# Clean; check for null values

for name, dataframe in all_df.items():
    null_counts = dataframe.isnull().sum()
    print(f"Null value counts in {name}:\n{null_counts}\n")
```

```python
# Check to see if data set has principal column, if it does replace any nulls with 'unlisted'
for name, dataframe in all_df.items():
    if 'Principal' in dataframe.columns:
        dataframe['Principal'] = dataframe['Principal'].fillna('Unlisted')

# Check for null values again
for name, dataframe in all_df.items():
    null_counts = dataframe.isnull().sum()
    print(f"Null value counts in {name}:\n{null_counts}\n")
```

```python
# Define a mapping of old column names to new column names for each dataset
column_name_changes = {
    'overview_17_18': {
        'Performance Index Percent': 'Performance Index Percent 2017-18'
    },
    'ratings_15_16': {
        'Performance Index Percent': 'Performance Index Percent 2015-16',
        'Letter Grade of Overall Value Added': 'Overall Value Added Grade 2015-16'
    },
    'ratings_16_17': {
        'Performance Index Percent': 'Performance Index Percent 2016-17'
    },
    'value_added_15_16': {
        'Overall Value Added Grade': 'Overall Value Added Grade 2015-16'
    },
    'value_added_16_17': {
        'Overall Value Added Grade': 'Overall Value Added Grade 2016-17'
    },
    'value_added_17_18': {
        'Overall Value Added Grade': 'Overall Value Added Grade 2017-18'
    }
}

# Apply the column name changes for each dataset
for name, changes in column_name_changes.items():
    dataframe = all_df[name]
    # Use rename with columns parameter and inplace=True to change the names directly in the DataFrame
    dataframe.rename(columns=changes, inplace=True)

# Print the new column names to verify the changes
for name, dataframe in all_df.items():
    print(f"Columns in {name}: {dataframe.columns.tolist()}\n")
```

```python
# Define IRN for Cleveland Metropolitan School District schools
# Use an integer for the IRN since the data is stored as int64
cmsd_irn = 43786  

# Create new dataframe with only CMSD schools
filtered_df = {}
for name, dataframe in all_df.items():
    if 'District IRN' in dataframe.columns:
        filtered_df[name] = dataframe.loc[dataframe['District IRN'] == cmsd_irn]
    else:
        print(f"'District IRN' column not found in {name}")

all_df = filtered_df


from IPython.display import display

for name, dataframe in all_df.items():
    print(f"Top 10 rows of {name}:")
    display(dataframe.head(10))
    print(f"Bottom 10 rows of {name}:")
    display(dataframe.tail(10))
```

```python
# Drop 'District IRN' and 'District Name' columns from each DataFrame
for name, dataframe in all_df.items():
    if 'District IRN' in dataframe.columns and 'District Name' in dataframe.columns:
        # inplace=True modifies the existing DataFrame without the need to assign it back
        dataframe.drop(['District IRN', 'District Name'], axis=1, inplace=True)

# Print the column names of each DataFrame to verify
for name, dataframe in all_df.items():
    print(f"Columns in {name}: {dataframe.columns.tolist()}\n")
```

```python
# Review head and tail of each dataset for accuracy
for name, dataframe in all_df.items():
    print(f"Top 10 rows of {name}:")
    display(dataframe.head(10))
    print(f"Bottom 10 rows of {name}:")
    display(dataframe.tail(10))
```

```python
# For careful and detailed data manipulation, dataframes are exported to one Excel sheet with an extra column added labeling what dataset each row was from
# Add a 'Source Dataset' column to each DataFrame
for name, dataframe in all_df.items():
    dataframe['Source Dataset'] = name 

# Concatenate all DataFrames into a single DataFrame
combined_df = pd.concat(all_df.values(), ignore_index=True)

# Export the combined DataFrame to an Excel file
output_file_path = r'C:\Users\sdnea\Documents\Data Science\CMSD Project\CMSD_Project_Combined_Data.xlsx'  # Specify your desired output file path
combined_df.to_excel(output_file_path, index=False)
```
